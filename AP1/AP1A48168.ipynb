Exercicio 3.1 - Slicing

```python
a=np.arange(0,10)
print(a)
print()
print(a[3:8:2]) # começa do index 3 até ao 8 saltando de 2 em 2
print()
print(a[-3:2:-1]) # -3 inicia do index 7 e vai até ao index 2 não inclusive andando de 1 em 1 de forma decrescente
print()
print(a[-3:]) # inicia do index -3 ou seja 7 e vai até ao final 
print()
print(a[-3:10]) # igual ao anterior mas aqui mete-se um limite até ao index 10
```

```python
A=np.array(([3,0,1],[0,4,1],[1,0,2]))
#A=np.matrix([[3,0,1],[0,4,1],[1,0,2]])
np.linalg.det(A) # determinante de A
np.linalg.inv(A) # inverso de A
val, Vec = np.linalg.eig(A) # Valores e vetores próprios de A
print('Val',val) #Valores de A
print('Vec',Vec) #vetores próprios de A
```

```python
#The sinc function , also called the "sampling function," is a function that arises frequently in signal processing and the theory of Fourier transforms. 
# The full name of the function is "sine cardinal," but it is commonly referred to by its abbreviation, "sinc." There are two definitions in common use.

#desenhar a função sinc(x)
x=np.linspace(-5,5,1000) #1000pts equi-espaçados entre [-5,5]
#arange necessita do passo, linspace necessita do número de passos
fx=np.sin(np.pi*x)/(np.pi*x) #fx  a função "sinc(x)"
plt.plot(x,fx,lw=1)
plt.axis([-5,5,-0.25,1.1])   #eixos 
plt.grid()                   #grelha
plt.xticks(np.arange(-5,6))  #intervalos no x = 1
plt.yticks(np.arange(-.25,1.25,.25))  #intervalos no y = 1/4
plt.text(.5,.75,r'$f(x)=\frac{\sin(\pi x)}{\pi x}$',fontsize=16)
plt.savefig('figs/L0AAex003a.png') #guardar  em ficheiro ".png"
plt.show()                            #(na directoria "../figs/")
```

Exercicio 6 

```python
#A distribuição normal ou Gaussiana é uma das mais importantes distribuições teóricas e práticas. Ela é muito utilizada na inferência estatística. 
# A função f(x) é uma curva simétrica, unimodal com forma de sino que, quando μ=0 e σ=1, ela descreve uma distribuição denominada Normal padrão ou Normal reduzida.

#gráfico de Gaussiana de média 0 (para dif.  variancias)
x=np.linspace(-5,5,1000) #1000pts equi-espaçados entre [-5,5]
sigma2=np.array([1./4,1./2,1.,2.]) #valores da variancia
plt.figure()
for s in sigma2:
    fx=1/np.sqrt(2*np.pi*s)*np.exp(-1./(2*s)*x**2)
    plt.plot(x,fx)
    strTmp=r'$\sigma^2=$%0.2f'%s
    plt.text(0,fx[500],strTmp,fontsize=16)
plt.axis([-5,5,0,1])
plt.xticks(np.arange(-5,6))    #intervalos no x =  1
plt.yticks(np.arange(0,1,.1))  #intervalos no y =  0.1 
plt.grid()
plt.savefig('figs/L0AAex003b.png') #guardar em ficheiro ".png"
plt.show()
```

```python
#representar versão amostrada de sinosóide
t=np.linspace(-1,2,1000) #1000pts equi-espaçados entre [-1,2]
fc=1 #frequ�ncia da sinosóide
x_t=np.cos(2*np.pi*fc*t) #fc Hz
N=10 #n� amostras segundo (N=fs, freq. de amostragem)
#tempo discreto (T=1/fs-> amostras de 0.1 em 0.1 segundos)
n=np.arange(-10,20) #[-1xN,2xN]
x_n=np.cos(2*np.pi*fc*n/N)
plt.plot(t,x_t,':b',lw=1)
plt.stem(n*1./N,x_n,'r')
plt.axis([-1,2,-1.05,1.05])  #eixos 
plt.grid(True)               #grelha
plt.xlabel(r'$t$ (segundos)'),plt.ylabel('amplitude')
plt.title(r'$x(t)=cos(2\pi t)\quadx[n]=cos(2\pi\frac{n}{10})$')
plt.xticks(np.arange(-1,2.1,.5))  #intervalos no x = 1/2
plt.yticks(np.arange(-1,1.1,.25))  #intervalos no y = 1/4
plt.savefig('figs/L0AAex003c.png') #guardar  em ficheiro ".png"
plt.show()                            #(na directoria "../figs/")
```

```python
#visualizar função f(x,y)=sin(x+x*y)*sin(y+x*y)/(x+x*y)/x
#criar grelha com função meshgrid
X,Y=np.meshgrid(np.linspace(-5,5.,50),np.linspace(-3,3.,50))
#X e Y arrays bi-dimensionais de 50x50
Z=np.sin(X+X*Y)*np.sin(Y+X*Y)/(Y+X*Y)/X
f1=plt.figure() #criar figura
ax=f1.add_subplot(111,projection='3d') #3D
ax.plot_wireframe(X,Y,Z,alpha=.3)
ax.contour(X,Y,Z,25,offset=Z.min()) #25=nº contornos
ax.elev=20   #posição da câmera:elevação
ax.azim=-150 #posição da câmera:azimute (em graus)
ax.set_xlabel('$x$',fontsize=6)
ax.set_ylabel('$y$',fontsize=6)
ax.set_zlabel('$f(x,y)$',fontsize=6)
plt.savefig('figs/L0AAex004a.png') #guardar em ficheiro ".png"
plt.show()
```
